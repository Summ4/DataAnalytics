```python
import pandas as pd
import warnings
from scipy.stats import norm, ttest_1samp, ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns



#N 1
data = pd.read_csv('LoanStatus.csv')

print(data.describe())


print(data.isnull().sum())

# N2 ა)
#სესხის აღების ალბათობის დათვლა
loan_status_counts = data['Loan_Status'].value_counts()
loan_status_probs = loan_status_counts / loan_status_counts.sum()

print("სესხის აღების ალბათობა:")
print(loan_status_probs)

# N2 ბ)
# ფილტრი, მაშინ როცა კლიენტს კარგი საკრედიტო ისტორია აქვს
good_credit_data = data[data['Credit_History'] == 1]


good_credit_prob = good_credit_data['Loan_Status'].value_counts().iloc[0] / len(good_credit_data)

print("სესხის აღების ალბათობა, მაშინ როცა კლიენტს კარგი საკრედიტო ისტორია აქვს:", good_credit_prob)

# N3 ა)
sns.distplot(data['ApplicantIncome'])
plt.show()

# N3 ბ)
plt.hist(data['LoanAmount'], bins=10)
plt.show()

# N4
mean_income = data['ApplicantIncome'].mean()
std_income = data['ApplicantIncome'].std()

# მომხარებლების პროცენტის გაოთვლა, რომელთაც შემოსავალი 2000 ევროზე ნაკლები აქვთ.
income_less_2000_prob = norm.cdf(2000, mean_income, std_income)
percentage_less_2000 = income_less_2000_prob * 100

print(f"მომხმარებელთა პროცენტი 2000 ევროზე ნაკლები შემოსავლით:", percentage_less_2000)

# N5 ა)
# boxplots  LoanAmount და Credit_History სვეტების მიხედვით
sns.boxplot(
    x = 'Credit_History',
    y = 'LoanAmount',
    showmeans=True,
    data=data
)
plt.show()

sns.boxplot(
    x = 'Married',
    y = 'ApplicantIncome',
    showmeans=True,
    data=data
)
plt.show()


# N6 ა)
mean_income = data['ApplicantIncome'].mean()
t_statistic, p_value = ttest_1samp(data['ApplicantIncome'], mean_income)

print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
import pandas as pd

# Task 1
data = pd.read_csv('LoanStatus.csv')
print(data.describe())

print(data.isnull().sum())

# Task 2a
loan_status_counts = data['Loan_Status'].value_counts()
loan_status_probs = loan_status_counts / loan_status_counts.sum()
print("Probability of getting a loan:")
print(loan_status_probs)

# Task 2b
good_credit_data = data[data['Credit_History'] == 1]
good_credit_prob = good_credit_data['Loan_Status'].value_counts().iloc[0] / len(good_credit_data)
print("Probability of getting a loan with good credit history:", good_credit_prob)

# Task 3a
sns.distplot(data['ApplicantIncome'])
plt.show()

# Task 3b
plt.hist(data['LoanAmount'], bins=10)
plt.show()

# Task 4
mean_income = data['ApplicantIncome'].mean()
std_income = data['ApplicantIncome'].std()
income_less_2000_prob = norm.cdf(2000, mean_income, std_income)
percentage_less_2000 = income_less_2000_prob * 100
print(f"Percentage of customers with income less than 2000 euros:", percentage_less_2000)

# Task 5a
sns.boxplot(
    x = 'Credit_History',
    y = 'LoanAmount',
    showmeans=True,
    data=data
)
plt.show()

sns.boxplot(
    x = 'Married',
    y = 'ApplicantIncome',
    showmeans=True,
    data=data
)
plt.show()

# Task 6a
mean_income = data['ApplicantIncome'].mean()
t_statistic, p_value = ttest_1samp(data['ApplicantIncome'], mean_income)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```
