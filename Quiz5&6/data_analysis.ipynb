```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

data = pd.read_csv('earthquakes_2023_global.csv')

# Task 1: Simple Linear Regression
X_simple = data[['mag']]
y_simple = data['depth']

X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(X_simple, y_simple, test_size=0.2, random_state=42)

imputer_simple = SimpleImputer(strategy='mean')
X_train_simple = imputer_simple.fit_transform(X_train_simple)
X_test_simple = imputer_simple.transform(X_test_simple)

model_simple = LinearRegression()

model_simple.fit(X_train_simple, y_train_simple)

predictions_simple = model_simple.predict(X_test_simple)

performance_simple = 1 - mean_squared_error(y_test_simple, predictions_simple) / y_test_simple.var()

plt.scatter(X_test_simple, y_test_simple, color='black')
plt.plot(X_test_simple, predictions_simple, color='blue', linewidth=3)
plt.xlabel('Magnitude')
plt.ylabel('Depth')
plt.title('Simple Linear Regression')
plt.show()

print(f"Task 1 - Simple Linear Regression პერფორმანსი: {performance_simple:.2f}")

# Task 2: Multiple Linear Regression
X_multiple = data[['mag', 'depth', 'horizontalError']]
y_multiple = data['rms']

X_train_multiple, X_test_multiple, y_train_multiple, y_test_multiple = train_test_split(X_multiple, y_multiple, test_size=0.2, random_state=42)

imputer_multiple = SimpleImputer(strategy='mean')
X_train_multiple = imputer_multiple.fit_transform(X_train_multiple)
X_test_multiple = imputer_multiple.transform(X_test_multiple)

model_multiple = LinearRegression()

model_multiple.fit(X_train_multiple, y_train_multiple)

predictions_multiple = model_multiple.predict(X_test_multiple)

performance_multiple = 1 - mean_squared_error(y_test_multiple, predictions_multiple) / y_test_multiple.var()

plt.scatter(range(len(X_test_dt)), y_test_dt, color='black', label='Actual')
plt.scatter(range(len(X_test_dt)), predictions_dt, color='blue', label='Predicted')
plt.xlabel('Data Points')
plt.ylabel('Depth')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()

print(f"Task 2 - Multiple Linear Regression პერფორმანსი: {performance_multiple:.2f}")

# Task 3: Decision Tree Regression
X_dt = data[['mag', 'depth', 'horizontalError']]
y_dt = data['rms']

X_train_dt, X_test_dt, y_train_dt, y_test_dt = train_test_split(X_dt, y_dt, test_size=0.2, random_state=42)

imputer_dt = SimpleImputer(strategy='mean')
X_train_dt = imputer_dt.fit_transform(X_train_dt)
X_test_dt = imputer_dt.transform(X_test_dt)

model_dt = DecisionTreeRegressor()

model_dt.fit(X_train_dt, y_train_dt)

predictions_dt = model_dt.predict(X_test_dt)

performance_dt = 1 - mean_squared_error(y_test_dt, predictions_dt) / y_test_dt.var()

plt.scatter(range(len(X_test_dt)), y_test_dt, color='black', label='Actual')
plt.scatter(range(len(X_test_dt)), predictions_dt, color='blue', label='Predicted')
plt.xlabel('Data Points')
plt.ylabel('RMS')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()

print(f"Task 3 - Decision Tree Regression პერფორმანსი: {performance_dt:.2f}")

# Task 4: Logistic Regression
X_logistic = data[['mag', 'depth', 'horizontalError']]
y_logistic = (data['mag'] > 4.0).astype(int)

X_train_logistic, X_test_logistic, y_train_logistic, y_test_logistic = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=42)

imputer_logistic = SimpleImputer(strategy='mean')
X_train_logistic = imputer_logistic.fit_transform(X_train_logistic)
X_test_logistic = imputer_logistic.transform(X_test_logistic)

model_logistic = LogisticRegression()

model_logistic.fit(X_train_logistic, y_train_logistic)

predictions_logistic = model_logistic.predict(X_test_logistic)

performance_logistic = accuracy_score(y_test_logistic, predictions_logistic)

print(f"Task 4 - Logistic Regression პერფორმანსი: {performance_logistic:.2f}")

# Task 5: Decision Tree Classification
X_tree_classification = data[['mag', 'depth', 'horizontalError']]
y_tree_classification = (data['mag'] > 4.0).astype(int)

X_train_tree_classification, X_test_tree_classification, y_train_tree_classification, y_test_tree_classification = train_test_split(X_tree_classification, y_tree_classification, test_size=0.2, random_state=42)

imputer_tree_classification = SimpleImputer(strategy='mean')
X_train_tree_classification = imputer_tree_classification.fit_transform(X_train_tree_classification)
X_test_tree_classification = imputer_tree_classification.transform(X_test_tree_classification)

model_tree_classification = DecisionTreeClassifier()

model_tree_classification.fit(X_train_tree_classification, y_train_tree_classification)

predictions_tree_classification = model_tree_classification.predict(X_test_tree_classification)

performance_tree_classification = accuracy_score(y_test_tree_classification, predictions_tree_classification)

print(f"Task 5 - Decision Tree Classification პერფორმანსი: {performance_tree_classification:.2f}")
```
